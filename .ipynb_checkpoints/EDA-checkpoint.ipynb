# Einblicke in den kompletten Datensatz mittels Zeitreihenanalyse, Scatterplots und Korrelationsmatrix

## 1. Zeitreihenanalyse (Liniendiagramm)

```python
# Line Plot -> doses admin / date
# Filtern nach gewünschten Ländern
country_list = ['Germany', 'Angola', 'US', 'Belarus', 'India']
filtered_df = df[df['Country'].isin(country_list)]

# X- und Y-Werte festlegen
x_date = filtered_df['Date']
y_doses = filtered_df['Doses_admin_per_100']

# Datumsformat anpassen
years = mdates.YearLocator()   # every year
months = mdates.MonthLocator()  # every month
years_fmt = mdates.DateFormatter('%Y-%m') #This is a format

# Figure und Achsen anlegen
fig, axes = plt.subplots(figsize=(20,10))

# Liniendiagramm plotten
sns.lineplot(x=x_date, y=y_doses,hue=filtered_df['Country'], marker = 'o')

# Achseneinteilung festlegen + Beschriftungen
axes.xaxis.set_major_locator(months)
plt.xticks(rotation = 'vertical')
plt.title('Zeitlicher Verlauf der verabreichten Impfdosen ab 22.Januar 2020', fontdict={'fontsize':30})

plt.show()
```

## 2. Scatterplots

```python
# Scatterplot mit Anzahl der verabreichten Impfdosen pro 100 Personen / Bruttonationaleinkommen pro Kopf 
# / relativer Anzahl der Todesfälle

fig = plt.gcf()

# aktuellstes Datum bekommen
date_max = df['Date'].max()

# Change seaborn plot size
fig.set_size_inches(15, 10)

# Filter -> zeige nur alle Daten vom letzten Datum
data_scatter = df[df['Date'] == date_max]

# Filter -> alle Werte "Doses_admin_per_100" größer als 0
data_scatter = data_scatter[data_scatter['Doses_admin_per_100'] > 0]

splot = sns.regplot(x=data_scatter["GNI_2019"], y=data_scatter["Doses_admin_per_100"], 
                    fit_reg=True, marker='o', scatter_kws={'s':data_scatter["Deaths_rel"]*1500},
                   line_kws={"color":"r","alpha":0.7,"lw":3})
splot.set(xscale="log")
splot.set_title('Anzahl Impfdosen / BNE pro Kopf / rel. Anzahl Tote',fontsize = 30)
```

```python
# Scatterplot mit Anzahl der verabreichten Impfdosen pro 100 Personen / Bruttonationaleinkommen pro Kopf 
# / relativer Anzahl bestätigter Corona-Fälle

fig = plt.gcf()

# Change seaborn plot size
fig.set_size_inches(15, 10)

# Filter -> zeige nur alle Daten vom letzten Datum
data_scatter_cases = df[df['Date'] == date_max]

# Filter -> alle Werte "Doses_admin_per_100" größer als 0
data_scatter_cases = data_scatter_cases[data_scatter_cases['Doses_admin_per_100'] > 0]

splot = sns.regplot(x=data_scatter_cases["GNI_2019"], y=data_scatter_cases["Doses_admin_per_100"], 
                    fit_reg=True, scatter_kws={'s':data_scatter_cases["Confirmed_Cases_rel"]*15},
                   line_kws={"color":"r","alpha":0.7,"lw":3})
splot.set(xscale="log")
splot.set_title('Anzahl Impfdosen / BNE pro Kopf / rel. Anzahl Covid-Fälle',fontsize = 30)
```

```python
# Scatterplot mit Anzahl der verabreichten Impfdosen pro 100 Personen / Bruttonationaleinkommen pro Kopf 
# / relativer Anzahl durchgeführter Corona-Tests

fig = plt.gcf()

# Change seaborn plot size
fig.set_size_inches(15, 10)

# Filter -> zeige nur alle Daten vom 01.05.2022
data_scatter_tests = df[df['Date'] == "2022-05-01"]

# Filter -> alle Werte "total_tests" größer als 0
data_scatter_tests = data_scatter_tests[data_scatter_tests['total_tests'] > 0]

splot = sns.regplot(x=data_scatter_tests["GNI_2019"], y=data_scatter_tests["Confirmed_Cases_rel"], 
                    fit_reg=True, scatter_kws={'s':(data_scatter_tests["total_tests_rel"])},
                   line_kws={"color":"r","alpha":0.7,"lw":3})
splot.set(xscale="log")
splot.set_title('Anzahl Impfdosen / BNE pro Kopf / rel. Anzahl Tests',fontsize = 30)
```

```python
# Anzahl der Länder mit Daten zu durchgeführten Tests
data_scatter_tests['Country'].nunique()
```

```python
# Korrelationsmatrix nach Pearson am 01.05.2022

df_corr = df[df['Date'] == "2022-05-01"]
df_corr = df_corr[{"GNI_2019", 'Doses_admin_per_100', 'Deaths_rel', "Confirmed_Cases_rel", "total_tests_rel", "GDP_pro_Kopf"}]

df_corr.corr(method="pearson").round(3).style.background_gradient(cmap='YlOrRd', high=0.5)
```
